**Pandas**

```python
import pandas as pd

# reading csv file w/ specific index, header, and skipped rows
df = pd.read_csv("../data/Salaries.csv", index_col="playerID", header=0, skiprows=[1])
df
```

```python
# selecting only atl & hou players with salaries above 1 million
trun_df = df[((df["teamID"] == "ATL") | (df["teamID"] ==  "HOU")) & (df["salary"] > 1000000)]
trun_df
```

```python
# describing the salaries of atl players
atl_df = df[df["teamID"] == "ATL"]
atl_df["salary"].describe()
```

**Numpy**

```python
import numpy as np

# creating 2d list and converting it to a numpy array
arr = [[0, 1], [2, 3], [4, 5]]
np_arr = np.array(arr)
np_arr
```

```python
# examining different numpy functions
print(np_arr.ndim)
print(np_arr.shape)
print(np_arr.size)
print(np_arr.dtype)
print(np_arr.itemsize)
print(type(np_arr))
```

```python
# examining dimension concepts
print(np_arr.reshape(2, 3))
print(np_arr.flatten())
```

```python
# understanding slice operation
print(np_arr[:1])
print(np_arr[::2])
```

```python
# examining ndarray operations
print(np_arr.mean())
print(np_arr.sum())
print(np_arr.min())
print(np_arr.max())

print(np.ones(5))
print(np.arange(5))
```

**Matplotlib**

```python
import matplotlib.pyplot as plt

# creating two 1d arrays
x = [0, 1, 2, 3, 4]
y = [0, 250, 500, 750, 1000]

plt.plot(x, y)
plt.title("y vs. x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

```python
plt.bar(x, y)
```
